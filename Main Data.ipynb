```python
# all plants url
url = 'https://plants.sc.egov.usda.gov/java/imageGallery?txtparm=&category=common&familycategory=all&duration=all&growthhabit=all&nativestatus=all&wetland=all&stateSelect=all&artist=all&copyright=all&imagetype=all&cite=all&location=all&viewsort=text&sort=sciname&submit2.x=59&submit2.y=6'
page = requests.get(url)
soup = BeautifulSoup(page.content,'html.parser')

plant_links, core_links, char_links = [], [], []
for link in soup.find_all('a'):
    if link.has_attr('href'):
        links = []
        links.append(link.attrs['href'])
        for link in links:
            if 'java/profile' in link:
                plant_links.append('https://plants.sc.egov.usda.gov'+link)
for link in plant_links:
    core_links.append(link.replace('java','core'))
    
# all plants characteristics url
url1 = 'https://plants.sc.egov.usda.gov/java/characteristics'
page1 = requests.get(url1)
soup1 = BeautifulSoup(page1.content,'html.parser')

for link in soup1.find_all('a'):
    if link.has_attr('href'):
        all_charry = []
        all_charry.append(link.attrs['href'])
    for link in all_charry:
        if 'profile?' in link:
            char_links.append(('https://plants.sc.egov.usda.gov/java/char'+link).replace('profile','Profile'))
            
print(('No. Plant Links',len(plant_links)), ('No. Core Characteristics Links',len(core_links)), ('No. Main Characteristics Links',len(char_links)))
```

```python
from bs4 import BeautifulSoup, NavigableString, Tag
```

```python
# url = 'https://plants.sc.egov.usda.gov/core/profile?symbol=ABGR4'

daters = []
for url in core_links:
    
# Make a GET request to fetch the raw HTML content
    html_content = requests.get(url).text

    # Parse HTML code for the entire site
    soup = BeautifulSoup(html_content, "lxml")

    for data in soup.find('table', {'class':'bordered'}):
        if isinstance(data, NavigableString):
            continue
        if isinstance(data, Tag):
            if 'General' not in data.text:
                if 'Characteristics' not in data.text:
                    dater = []
                    dater.append(data.text.replace('\t','').replace('\n','').replace(' \xa0 ',' '))
                    daters.append(dater)

# for data in soup.find('table', {'class':'bordered'}):
#     for dat in soup.find("table", attrs={"class": "bordered"}):
#         print(dat)
```
